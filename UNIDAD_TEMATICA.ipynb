## Igualdades: Identidades y ecuaciones

### Monomios y binomios

---
Un monomio consta de un sólo término formado por un producto

Ejemplos:

$$x^2$$
$$yx^2$$
$$m^3x^2$$

---
Un binomio consta de una suma o diferencia de dos monomios

Ejemplos:

$$x^2-y^2$$
$$a+b$$


### Igualdad algebraica

---
Una igualdad algebraica se obtiene al igualar dos expresiones algebraicas

$$2x + 5x = 7x$$
$$9y - 5y = 4y$$
$$(x + 1)^2 = x^2 + 2x + 1$$

#### Ejemplo

---
Cuando una identidad algebraica se cumple siempre se llama identidad 


Una identidad es una igualdad algebraica que se verifica para cualquier valor de las variables que aparecen

---
Dada esta igualdad notable 

$$(a + b)^2 = a^2 + 2ab + b^2$$

Se puede comprobar que es cierta para a = 5 y b = 1

$$ (5 + 1)^2 = 5^2 + 2 \cdot 5 \cdot 1 + 1^2 = 36 $$


ACTIVIDAD: Dada esta igualdad notable comprueba que es cierta para $a = 2$ y $b = 3$


---
Dada esta igualdad notable 

$$(a - b)^2 = a^2 - 2ab + b^2$$

Se puede comprobar que es cierta para a = 5 y b = 1

$$ (5 - 1)^2 = 5^2 - 2 \cdot 5 \cdot 1 + 1^2 = 16 $$

#### Ten en cuenta 

---
Las fórmulas que se usan en matemáticas y otras áreas también son igualdades algebraicas. Por ejemplo:
    
la base $b$ de un triángulo es el doble que su altura $h$

$$b = 2h$$

El peso $p$ de una persona en la Tierra se obtiene multiplicando su masa $m$ por la gravedad $9.8 m/s^2$

En cambio otras igualdades algebraicas no se verifican para cualquier valor de las variables y se denominan ecuación

### Definición de ecuación


Una ecuación es una igualdad algebraica que se verifica solamente en ctos valores de las variables (o incógnitas). Estos valores son las soluciones de la ecuación

El grado de la ecuación es el mayor grado de los monomios que contiene

#### Ejemplo


la suma de los dos ortoedros que forman este cuerpo es de $44 cm^3$

### Soluciones de una ecuación 


Resolver una ecuación es determinar el valor o valores que tiene que tomar la incógnita para que se cumpla la igualdad. Estos valores son las soluciones de la ecuación

Una ecuación puede tener una solución, varias, o infinitas o no tener ninguna

#### Ejemplos


La ecuación $3x - 5 = 7$ tiene una única solución, x = 4, ya que solo ese valor cumple la igualdad


La ecuacion $x + y = 3$ tiene infinitas soluciones ya que para cadad valor de x hay siempre un valor de y que cumple la igualdad


La ecuación $x + 1 = x + 2$ no tiene solución ya que si a un número se le suma una unidad siempre será menor que si le sumen dos unidades


La ecuación $x^2 = 9$ tiene dos soluciones ya que hay dos valores de x a los cuales el cuadrado vale 9, $x = -3$ y $x = 3$

### Reglas de la suma y del producto

Para la resolución de ecuacoines se aplican técnicas que permitan determinar otra ecuación que sea más sencilla utilizando la regla de la suma y del producto

#### Regla de la suma

Si en una ecuación se suma o se resta el mismo numero o la misma expresión algebraica en los dos miembros se obtiene una expresion equivalente 

$$3x + 6 = x +16$$
$$3x + 6 \color{red}{-6} = x +16 \color{red}{-6}$$
$$3x = x + 10$$
$$3x \color{blue}{-x} = x + 10 \color{blue}{-x}$$
$$2x = 10 $$

#### Regla del producto


Si en una ecuación se multiplica o se divide por un mismo factor (distinto a cero) en los dos miembros se obtiene una expresion equivalente 


Aplica la regla del producto para aislar el valor de x en la expresión $2x = 10$


$$2x = 10$$

$$2x  \color{blue}{ /2}= 10  \color{blue}{ /2}$$

$$x = 5$$

$$ \color{red}{x^2} -x = \color{red}{x^2} + 3x-5 $$

$$ -x=3x-5 $$

### Resolución de ecuaciones de primer grado con una incógnita

Para resolver una ecuación de este tipo, se aisla la incógnita siguiendo los pasos siguientes:
    
    1) Se opera para suprimir los parentesis
    
    2) Se opera para eliminar los denominadores
    
    3) Se simplifican los términos que se puedan
    
    4) Se aplican las reglas de la suma y del producto
